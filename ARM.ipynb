# 패키지 설정

```python
from mlxtend.frequent_patterns import apriori, association_rules # apriori 알고리즘
from mlxtend.preprocessing import TransactionEncoder # 연관분석용 데이터 파일 변환
import pandas as pd
import numpy as np
```

# ARM 수행하기

## 1. 데이터 작성

```python
# 장바구니 데이터
tran =[['bread', 'milk'],
       ['bread', 'diaper', 'beer' 'eggs'],
       ['milk', 'diaper', 'beer', 'coke'],
       ['bread', 'milk', 'diaper', 'beer'],
       ['bread', 'milk', 'diaper', 'coke']]
```

```python
# 데이터 세트를 연관분석이 가능한 배열로 변환 (옅은 아이템, 행은 거래건, 셀은 거래 유무)
te = TransactionEncoder()
# fit 함수로 데이터셋은 고유 레이블을 갖고, transform 함수로 파이썬 리스트를 one-hot-encoding된 numpy 배열로 변환
tran_ar = te.fit(tran).transform(tran)
print(tran_ar)
```

```python
# 열이름
print(te.columns_)
```

```python
# 각 상품명을 열이름으로 하는 데이터 프레임으로 변환
df = pd.DataFrame(tran_ar, columns = te.columns_)
print(df)
```

## 2. 탐색적 데이터 분석

```python
# 각 상품별 거래 빈도
freq = df.sum().to_frame('Frequency')
# 빈도 역순으로 정렬
freq_sort = freq.sort_values('Frequency', ascending= False)
print(freq_sort)
```

```python
# 지지도 계산 = 각 상품의 거래 빈도 / 총 거래 수
freq_sort['Support'] = freq_sort['Frequency'] /len(df)
print(freq_sort)
```

```python
# 상품별 거래 빈도
freq_sort['Frequency'].plot(kind='bar', title='FrequentItems');
```

```python
# 상품 지지도의 히스토그램
freq_sort['Support'].hist(bins=6);
```

```python
#최소 지지도 0.4 반발항목집합
freq_items = apriori(df, min_support=0.4, use_colnames=True)
print(freq_items)
```

```python
# 빈발항목잡합에 대하여 최소신뢰도 0.7의 연관규칙이 아님
rules = association_rules(freq_items,
                          metric = 'confidence',
                          min_threshold = 0.7)
rules
```

```python
# Lift가 1보다 큰 양의 관계를 만족하는 연관규칙들을 추출하고,
rules = rules[(rules['lift']>1)]
# 신뢰도의 내림차순으로 정렬
rules.sort_values(by= 'confidence', ascending = False)
```

```python
# 조건부에 'diaper'가 있는 규칙 
rules[rules.antecedents.apply(lambda x:'diaper' in x)]
```

```python
# 결론부에 'milk'가 있는 규칙
rules[rules.consequents.apply(lambda x:'milk' in x)]
```

# 실전예제 : 온라인 유통 상품의 연관성 분석

## 1. 데이터 파일 준비
* UCI Machine learning Repository ([https://archive.ics.uci.edu/ml/index.php](https://archive.ics.uci.edu/ml/index.php)) --> view all data sets

  * 이 데이터에는 결측치가 없음
  * InvoiceNo(거래번호), StockCode(상품코드), Description(상품명), Quantity(주문량), InvoiceDate(거래일시)
  * UnitsPrice(단가), CustomerID(고객ID), Country(고객의 거주국가)

```python
# 데이터 파일 불러오기
df = pd.read_excel('Online Retail.xlsx')
df
```

```python
df = df[['InvoiceNo', 'Description', 'Quantity', 'UnitPrice']]
```

## 2. 탐색적 데이터 분석

```python
# 주문량이 0이하인 거래가 있는지 확인
df[df['Quantity'] <= 0]
```

```python
# 주문량이 0보다 큰 거래만 추출
df = df[df['Quantity'] > 0]
df
```

```python
# 거래별 상품명별 수량 합계 (하나의 거래에 동일한 품목이 있는 경우가 있어 그룹별로 나눔)
df = df.groupby(['InvoiceNo', 'Description'])['Quantity'].sum()
df
```

```python
# 데이터프레임의 구조변경
## 1열 TnvoiceNo는 행으로, 2열 Description(상품명)은 열로, 3열 수량은 행렬의
df = df.unstack()
df
```

```python
# 각 셀의 수량이 1이상이면 1로, NAN이면 0으로 변환
## 함수 : 전달되는 x값들을 조건에 따라 0, 1로 변환
def encode_units(x):
  if x >=1 : return 1
  else: return 0
# df 데이터세트에 대해 encode_units 함수 적용
df = df.applymap(encode_units)
df
```

```python
# 각 열별로 합하고 내림차순으로 정렬하여 가장 많은 수량으로 거래된 상품명 확인하기
# 아이템 빈도 및 항목별 설정
freq = df.sum(axis=0).to_frame('Frequancy')
# 빈도의 내림차순 정렬
freq_sort = freq.sort_values('Frequancy', ascending= False)
freq_sort
```

```python
# 빈도수가 1인 항목 제외
freq_sort = freq_sort[freq_sort['Frequancy']>1]
freq_sort
```

```python
# 단일 상품별 지지도
freq_sort['Support'] = freq_sort['Frequancy'] / len(freq_sort)
freq_sort
```

```python
# 지지도 분포
freq_sort['Support'].hist()
```

## 3. 연관규칙 생성

```python
# 최소 지지도 0.01 (메모리 오류 발생시 0.05부터 조정할 것)
freq_items = apriori(df, min_support=0.02, use_colnames=True)
freq_items
```

```python
# 최소 신뢰도 0.6으로 연관규칙 생성
rules = association_rules(freq_items, metric='confidence', min_threshold=0)
rules
```

```python
# 향상도가 1보다 큰 연관규칙들을 추출하고, 신뢰도의 내림차순으로 정렬
rules = rules[rules['lift'] > 1]
rules.sort_values(by='confidence', ascending=False)
```
